```python
import pandas as pd
```

```python
fg = pd.read_csv("netflix_titles.csv")
```

```python
fg['director'] = fg['director'].fillna("Unknown")
fg['cast'] = fg['cast'].fillna("Unknown")
fg['country'] = fg['country'].fillna("Unknown")
```

```python
fg.dropna(subset=['date_added','rating','duration'], inplace=True)
```

```python
fg['date_added'] = pd.to_datetime(fg['date_added'].astype(str).str.strip(), errors='coerce')
```

```python
fg.info()
```

```python
fg.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
# Number of titles(increased/decreased) in proportion to the years
yearly_counts = fg['date_added'].dt.year.value_counts().sort_index()
plt.style.use('grayscale')
plt.plot(yearly_counts.index, yearly_counts.values, 'bo--')
plt.xlabel("Year Document Added")
plt.ylabel("Number of Titles")
plt.title("Netflix Content Growth Overtime")
plt.grid(True)
plt.tight_layout()
plt.savefig("Netflix_Content_Growth_Overtime2.png")
```

```python
top_countries = fg['country'].value_counts().head(10)
plt.style.use("ggplot")
plt.bar(top_countries.index, top_countries.values, color = "forestgreen", edgecolor="black")
plt.xlabel("Country Name")
plt.ylabel("Number of Titles")
plt.title("Top 10 Countries by Netflix Content")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("Bar_chart_Top_10_Countries_Netflix.png")
```
